```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python
#plotting the ecdf to confirm if data is distributed 
x,y = ecdf(df1['Age'])
_ = plt.plot(x, y, marker='.', linestyle='none')

# Label the axes
_ = plt.xlabel('Age')
_ = plt.ylabel('ECDF')


# Display the plot
plt.show()

sns.distplot(df1['Age'], hist=True, kde=True, color = 'darkblue', 
             bins=48, hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 2})
```

```python
#analysing the difference in negative perception of mental health discussion with the employers in workplace.
consequences_data = df1[['mental_health_consequence', 'phys_health_consequence']]
consequences_melted = consequences_data.melt(var_name='Health Issue', value_name='Perceived Consequence')
plt.figure(figsize=(6, 4))
sns.countplot(x='Health Issue', hue='Perceived Consequence', data=consequences_melted)
plt.title('Perceived Negative Consequences of Discussing Health Issues at Work')
plt.xlabel('Type of Health Issue')
plt.ylabel('Number of Respondents')
plt.legend(title='Negative Consequence')
plt.show()
```

This clearly indicates that a majority of people do not believe discussing physical health has negative consequences in 
the workplace. However, many individuals express uncertainty regarding the impact of discussing mental health concerns 
in the same setting. This is evident from the increased proportion of 'maybe' responses compared to discussions about 
physical health, suggesting a decrease in confidence. It underscores the prevailing stigma around mental health.

Employers can leverage this data representation to discern disparities between perceptions of physical and mental health 
and to foster an environment that encourages employees to bridge these gaps by offering relevant resources
